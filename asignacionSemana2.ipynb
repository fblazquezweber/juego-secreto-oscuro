```python
opciones = ["Consultar","Agregar" ,"Modificar", "Eliminar","Listar", "Salir"]
modificaciones = ["Titulo","Autor" ,"Editorial", "Genero","Precio", "Unidades", "Descripcion", "Salir"]
titulos = ["Desastre", "La Casa", "El Detective"]
autores = ["Luis Brito", "Jose Perez", "Pirlo Camonero"]
editoriales = ["La Ceiba", "Los Alpes", "Editorial Risueña"]
generos = ["Misterio", "Romance", "Comedia"]
precios = [250, 100, 150]
unidades = [10, 5, 6]
descripciones = [
    "Intrigante",
    "Dramatico",
    "Muchas risas"
    ]

def mostrarMenu():
  print('------Menu opciones------')
  for i in range(len(opciones)):
    print(f'{i+1}. {opciones[i]}')
  print('-------------------------')

def mostrarModificar():
  print('------Menu opciones------')
  for i in range(len(modificaciones)):
    print(f'{i+1}. {modificaciones[i]}')
  print('-------------------------')

def buscarLibro():
  buscar = input('Ingrese el titulo del libro a buscar: ')
  try:
    indice = titulos.index(buscar)
    print(f'Titulo: {titulos[indice]}')
    print(f'Autor: {autores[indice]}')
    print(f'Editorial: {editoriales[indice]}')
    print(f'Genero: {generos[indice]}')
    print(f'Precio: {precios[indice]}')
    print(f'Unidades: {unidades[indice]}')
    print(f'Descripcion: {descripciones[indice]}')
    print(f'El libro se encuentra en el estante número {indice}')
  except ValueError:
    print(f'El libro {buscar} se encuentra agotado')

def agregarLibro():
  titulos.append(input('Ingrese el titulo del libro insertar: '))
  autores.append(input('Ingrese el autor del libro insertar: '))
  editoriales.append(input('Ingrese la editorial del libro insertar: '))
  generos.append(input('Ingrese el genero del libro insertar: '))
  precios.append(int(input('Ingrese el precio del libro insertar: ')))
  unidades.append(int(input('Ingrese la cantidad disponible para este titulo: ')))
  descripciones.append(input('Ingrese la descripcion del libro insertar: '))

def modificarLibro():
   buscar = input('Ingrese el titulo del libro a modificar: ')
   try:
    while True:
      indice = titulos.index(buscar)
      mostrarModificar()
      opcion = int(input('Ingrese una opcion: '))
      if opcion == 1:
        titulos[indice] = input('Ingrese el nuevo titulo: ')
      elif opcion == 2:
        autores[indice] = input('Ingrese el nuevo autor: ')
      elif opcion == 3:
        editoriales[indice] = input('Ingrese la nueva editorial: ')
      elif opcion == 4:
        generos[indice] = input('Ingrese el nuevo genero: ')
      elif opcion == 5:
        precios[indice] = int(input('Ingrese el nuevo precio: '))
      elif opcion == 6:
        unidades[indice] = int(input('Ingrese la nueva cantidad: '))
      elif opcion == 7:
        descripciones[indice] = input('Ingrese la nueva descripcion: ')
      elif opcion == 8:
        break
    else:
      print('Opcion no valida')
   except ValueError:
    print(f'El libro {buscar} se encuentra agotado')

def eliminarLibro():
   eliminar = input('Ingrese el titulo del libro a eliminar: ')
   try:
      indice = titulos.index(eliminar)
      titulos.pop(indice)
      autores.pop(indice)
      editoriales.pop(indice)
      generos.pop(indice)
      precios.pop(indice)
      unidades.pop(indice)
      descripciones.pop(indice)
   except ValueError:
    print(f'El libro {eliminar} se encuentra agotado')

def listarLibros():
  for i in range(len(titulos)):
    print(f'Titulo: {titulos[i]}')

while True:
  mostrarMenu()
  opcion = int(input('Ingrese una opcion: '))
  if opcion == 1:
    buscarLibro()
  elif opcion == 2:
    agregarLibro()
  elif opcion == 3:
    modificarLibro()
  elif opcion == 4:
    eliminarLibro()
  elif opcion == 5:
    listarLibros()
  elif opcion == 6:
    break
  else:
    print('Opcion no valida')
```
